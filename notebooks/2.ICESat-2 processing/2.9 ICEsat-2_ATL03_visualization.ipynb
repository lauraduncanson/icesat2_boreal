# 2.9 ICESat-2 photon data visualization

```r
# Libraries and functions

# load packages using pacman
library(pacman)
p_load(readxl)

coloRclass<-function(data, col=c("gray","red","blue","green")){
  
  cols<-1:nrow(data)
  for ( i in 1:nrow(data)){
    #print(i)
    if ( data$Classification[i]==0){cols[i]<-col[1]}
    if ( data$Classification[i]==1){cols[i]<-col[2]}
    if ( data$Classification[i]==2){cols[i]<-col[3]}
    if ( data$Classification[i]==3){cols[i]<-col[4]}
    
  }
  
  data$cols<-cols
  return(data)
}
```

```r
# import files

ws<-"~/csilva/ATL03_visualization/"
LaSelva4<-coloRclass(read_excel(file.path(ws,"LaSelva_repeat7_track1_beam7.xls")))
Lope1<-coloRclass(read_excel(file.path(ws,"Lope_repeat1_track1_beam1.xls")), col=c("gray","brown","green","forestgreen"))
```

```r
# visualization
plot(LaSelva4$Longitude,LaSelva4$Height, col=LaSelva4$cols, pch=16, cex=0.75,font.lab=2, font.axis=2,cex.axis=1.5, xlab="Longitude (º)", ylab="Elevation (m)", xlim=c(10.3975,10.449))
legend(10.43,280, legend=c("Raw data","Canopy","Canopy top","Ground"), col=c("gray","blue","green","red"), bty="n", pch=19, text.font = 2, cex=1.5)
```

```r
# visualization
plot(Lope1$Longitude,Lope1$Height, col=Lope1$cols, pch=16, cex=0.75,font.lab=2, font.axis=2,cex.axis=1.5, xlab="Longitude (º)", ylab="Elevation (m)")
legend("topright", legend=c("Raw data","Canopy","Canopy top","Ground"), col=c("gray","green","forestgreen","brown"), bty="n", pch=19, text.font = 2, cex=1.5)
```
